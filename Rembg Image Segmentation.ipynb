```python
import cv2
from random import shuffle
import matplotlib.pyplot as plt
from rembg import remove, new_session
from glob import glob
```

```python
## All images shuffled
session = new_session()
data = []
files = glob('dataset/*/*.png')
shuffle(files)

for file in files:
    input_path = str(file)
    
    bx = cv2.imread(input_path)
    output = remove(bx, session=session)
    data.append(output)

plt.figure(figsize=(12,12)) 
for i in range(len(data)):
    ax = plt.subplot(len(data)//5+1,1,i+1)
    plt.imshow(data[i])
    #plt.title(f"Original {class_names[label_batch[i]]}")
    plt.axis("off")
```
