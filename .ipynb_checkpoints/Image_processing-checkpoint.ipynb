```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 
from skimage import io, measure
```

```python
lettuce = cv2.imread("C:/Users/Patrick Formales/Desktop/fyp/dataset/FN/fn1.png")
lettucergb = cv2.cvtColor(lettuce, cv2.COLOR_BGR2RGB)
plt.imshow(lettucergb)
plt.show()

lettucehsv = cv2.cvtColor(lettucergb, cv2.COLOR_RGB2HSV)
plt.imshow(lettucehsv)
plt.show()

```

```python
mask = cv2.inRange(lettucehsv, (30, 0, 0), (86,255,255))
plt.imshow(mask)
plt.show()

mask_black = cv2.inRange(lettucehsv, (0,0,0), (180,255,40))
plt.imshow(mask_black)

mask_total = mask + mask_black
plt.imshow(mask_total)
```

```python
from scipy import ndimage as nd 
closed_mask = nd.binary_closing(mask, np.ones((15,15)))
plt.imshow(closed_mask)
```

```python
label_image = measure.label(closed_mask)
plt.imshow(label_image)
```

```python
from skimage.color import label2rgb 
image_label_overlay = label2rgb(label_image, image = lettucergb)
plt.imshow(image_label_overlay)
```

```python
segmented = cv2.bitwise_and(lettucergb, lettucergb, mask=mask_total)
plt.imshow(segmented)
```
